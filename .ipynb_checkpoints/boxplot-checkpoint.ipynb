```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import MultipleLocator
import numpy as np

figure(num=None, figsize=(7, 5), dpi=80, facecolor='w', edgecolor='k')
ml = MultipleLocator(10)
meanlineprops = dict(linestyle='--', linewidth=1.5, color='blue')

plt.plot(range(290,300),range(0,10))
B=[1,2,5,6,1,2,9,7,5,4,10,12,4,0]
C=np.array(B)*1.5
Winter_BOX=plt.boxplot([B,C],meanprops=meanlineprops,showmeans=True, meanline=True,sym='',positions=[293,294])


   
plt.ylabel('ERA5 - PCIC [mm]',fontsize=15)
plt.ylim(-5,20)
plt.xlim(290,300)
plt.yticks(fontsize=12)
plt.xticks(fontsize=15)
plt.grid(which='both',axis='y')
plt.title('C. Low',fontsize=15)
```

```python
def make_gradients(path, path_out, dh_dt, topo, vel, msk, n_min,num):
    num=num
    
    glacier = ['conrad','illecillewaet','kokanee']
    glacier= glacier[num]

    path = '/home/pelto/Desktop/lidar_cbt_analysis/' + glacier + '/'
    
    
    NAME = 'vz', 'Z', 'slope', 'ipr'  #'Depth' 'v'
    bin_size = 50.
    
    
    ipr_list= ['/conrad_obs_all_50m.tif','/illec_gpr_all_50m.tif','/kokanee_gpr_final_50m.tif']
    ipr= '/home/pelto/GIS/GPR/ComparisonData/' + glacier + ipr_list[num]
    slope_list = ['ConradSlope160912_20m.tif','Illec_slope180426_10m.tif','Kokanee150912.tif']
    slope = '/home/pelto/GIS/DEM/Slope/'+ slope_list[num]
    vdir=['/home/pelto/Desktop/velocity_mapping/Conrad_DEMs/spm2/3m/','/home/pelto/Desktop/velocity_mapping/illecillewaet/3m/',
         '/home/pelto/Desktop/velocity_mapping/kokanee/bedem/3m/']
    vdir=vdir[num]
    if num==0:
        VX = vel[:-5] + 'x.tif' ; VY = vel ;  VM = vel[:-5] + 'm.tif'  
    elif num==1:
        VX = vel ; VY = vel[:-9]+ 'y.tif' ;  VM = vel[:-9]+ 'y.tif'  
    else:
        VX = vel ; VY = vel[:-5]+ 'y.tif' ;  VM = vel[:-5]+ 'y.tif'  
        
    vm = salem.open_xr_dataset(VM);vx = salem.open_xr_dataset(VX);vy = salem.open_xr_dataset(VY)
    slope=salem.open_xr_dataset(slope); ipr=salem.open_xr_dataset(ipr)
    vm = vm.to_array(name='vm');#vx = vx.to_array(name='vx');vy = vy.to_array(name='vy')
    vx_reproj = vm.salem.transform(vx);vy_reproj = vm.salem.transform(vy);ipr_reproj = vm.salem.transform(ipr);
    vx = vx_reproj.to_array(name='vx');vy = vy_reproj.to_array(name='vy');slope_reproj = vm.salem.transform(slope)
    ipr = ipr_reproj.to_array(name='ipr');slope = slope_reproj.to_array(name='slope')
#     vz=vy.copy(name='vz')
    vz=np.sqrt(vx.data**2 + vy.data**2 )
    dem = salem.open_xr_dataset(path+topo);dem_reproj = vm.salem.transform(dem); dem = dem_reproj.to_array(name='dem')
   
    dh = salem.open_xr_dataset(path+dh_dt); dh_reproj = vm.salem.transform(dh); dh = dh_reproj.to_array(name='dh')
    msk = salem.open_xr_dataset(path+msk);msk_reproj = vm.salem.transform(msk);msk = msk_reproj.to_array(name='msk')
    
    vz[msk.data<1.0] = np.nan; dem.data[msk.data<1.0]=np.nan; slope.data[msk.data<1.0]=np.nan##remove data off-ice
    vz[vz==0.0]=np.nan; vz[vz<0.1]=np.nan; dem.data[dem.data<1000] = np.nan; ipr.data[ipr.data<0.1]=np.nan

    grid = vm.salem.grid  ##full view
    sm = Map(grid, countries=False)
    sm.set_lonlat_contours(interval=0)
    sm.set_scale_bar()
    sm.set_data(ipr)
    sm.visualize()
    
    print('Average velocity:', np.round(np.nanmean(vz),2), '+/-',np.round(np.nanstd(vz),2), 'm/yr')
    print('Average Depth:', np.round(np.nanmean(ipr),2), '+/-',np.round(np.nanstd(ipr),2), 'm/yr')
    print('Average Slope:', np.round(np.nanmean(slope),2), '+/-',np.round(np.nanstd(slope),2), 'm/yr')
    print('Median Elevation:', np.round(np.nanmean(dem),2), '+/-',np.round(np.nanstd(dem),2), 'm/yr')
    
    # convert 2-d elevation and SWE arrays to 1-D vectors using ravel function
    Zvec = np.ravel(dem);Velvec = np.ravel(vz);Slpvec=np.ravel(slope);iprvec=np.ravel(ipr)  
    
    bins = range(1850,3350,50)
       
    # create pandas dataframe for elevation and height change
    df = pd.DataFrame({"Z":Zvec, "vz":Velvec,  "slope":Slpvec,"ipr":iprvec,}) #"Depth":Depthvec,
  
    # classify the elevation data into 100 metre bins and add to the data frame 
    df['Zbins'] = pd.cut(df['Z'], bins, labels= np.arange(1875, 3325, 50))
    # get some statistics from the data # this will be series which can be a pain to work with 
    # so convert to dataframes
    
    P=[]
    for n in NAME: #'v' 'Depth'
    
        mean = df.groupby(by='Zbins')[n].mean()
        median = df.groupby(by='Zbins')[n].median()
#         iqr    = df.groupby(by='Zbins')[n].quantile(0.75) - df.groupby(by='Zbins')[n].quantile(0.25)
        MAX = df.groupby(by='Zbins')[n].max()
        count  = df.groupby(by='Zbins')[n].count()

        ## convert series to pandas data frames instead of series
#         if 
        if n=='Z':
            df_dem= pd.DataFrame(list(zip(mean,count)), columns=['dem_mean','dem_count'])
#             df_dem.reset_index(inplace=True)
            df_dem['Zbins'] = np.arange(1875, 3325, 50) 
            
        else:
            df_loop= pd.DataFrame(list(zip(mean,median, MAX,count)), columns=[n+'_mean',n+'_median',n+'_max', n+'_count'])
            #             df_vz= pd.DataFrame(list(zip(mean,median, iqr,count)), columns=['vz_mean','vz_median','vz_iqr', 'vz_count'])
            P.append(df_loop)
    ## returns data frame 
    DF =  pd.concat([df_dem, P[0], P[1], P[2]], axis=1)
    DF['VZcoverage'] = DF.vz_count / DF.dem_count 
    return DF
```

```python
def make_gradients_2(path, path_out, topo, msk, n_min,num):
    num=num
    
    glacier = ['nordic','zillmer']
    glacier= glacier[num]

    path = '/home/pelto/Desktop/lidar_cbt_analysis/' + glacier + '/'
    
    
    NAME = 'Z', 'slope', 'ipr'  #'Depth' 'v'
    bin_size = 50.
        
    ipr_list= ['/nordic_gpr_final_50m.tif','/zillmer_gpr_final_50m.tif']
    ipr= '/home/pelto/GIS/GPR/ComparisonData/' + glacier + ipr_list[num]
    slope_list = ['140911_nordic_slope_20m.tif','160914_zillmer_slope_20m.tif']
    slope = path + slope_list[num]
    slope=salem.open_xr_dataset(slope); ipr=salem.open_xr_dataset(ipr)
    dem = salem.open_xr_dataset(path+topo); dem = dem.to_array(name='dem')
    ipr_reproj = dem.salem.transform(ipr);   slope_reproj = dem.salem.transform(slope)
    ipr = ipr_reproj.to_array(name='ipr');slope = slope_reproj.to_array(name='slope')
   
    msk = salem.open_xr_dataset(path+msk);msk_reproj = dem.salem.transform(msk);msk = msk_reproj.to_array(name='msk')
    
    dem.data[msk.data<1.0]=np.nan;slope.data[msk.data<1.0]=np.nan ##remove data off-ice
    dem.data[dem.data<1000] = np.nan; ipr.data[ipr.data<0.1]=np.nan

    grid = dem.salem.grid  ##full view
    sm = Map(grid, countries=False)
    sm.set_lonlat_contours(interval=0)
    sm.set_scale_bar()
    sm.set_data(slope)
    sm.visualize()
    
    print('Average Depth:', np.round(np.nanmean(ipr),2), '+/-',np.round(np.nanstd(ipr),2), 'm/yr')
    print('Average Slope:', np.round(np.nanmean(slope),2), '+/-',np.round(np.nanstd(slope),2), 'm/yr')    
    print('Median Elevation:', np.round(np.nanmean(dem),2), '+/-',np.round(np.nanstd(dem),2), 'm/yr')
    # convert 2-d elevation and SWE arrays to 1-D vectors using ravel function
    Zvec = np.ravel(dem);Slpvec=np.ravel(slope);iprvec=np.ravel(ipr)  
    
    bins = range(1850,3350,50)
       
    # create pandas dataframe for elevation and height change
    df = pd.DataFrame({"Z":Zvec, "slope":Slpvec,"ipr":iprvec,}) #"Depth":Depthvec,
  
    # classify the elevation data into 100 metre bins and add to the data frame 
    df['Zbins'] = pd.cut(df['Z'], bins, labels= np.arange(1875, 3325, 50))
    # get some statistics from the data # this will be series which can be a pain to work with 
    # so convert to dataframes
    
    P=[]
    for n in NAME: #'v' 'Depth'
    
        mean = df.groupby(by='Zbins')[n].mean()
        median = df.groupby(by='Zbins')[n].median()
#         iqr    = df.groupby(by='Zbins')[n].quantile(0.75) - df.groupby(by='Zbins')[n].quantile(0.25)
        MAX = df.groupby(by='Zbins')[n].max()
        count  = df.groupby(by='Zbins')[n].count()

        ## convert series to pandas data frames instead of series
#         if 
        if n=='Z':
            df_dem= pd.DataFrame(list(zip(mean,count)), columns=['dem_mean','dem_count'])
#             df_dem.reset_index(inplace=True)
            df_dem['Zbins'] = np.arange(1875, 3325, 50) 
            
        else:
            df_loop= pd.DataFrame(list(zip(mean,median, MAX,count)), columns=[n+'_mean',n+'_median',n+'_max', n+'_count'])
            P.append(df_loop)
    ## returns data frame 
    DF =  pd.concat([df_dem, P[0], P[1]], axis=1)
#     DF['VZcoverage'] = DF.vz_count / DF.dem_count 
    return DF
```
